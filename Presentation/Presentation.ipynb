### Time stretching is the process of changing the speed or duration of an audio signal without affecting its pitch. 

### Pitch scaling is the opposite: the process of changing the pitch without affecting the speed.

### Ways to increase playback speed of the audio:
### Increasing the sample rate

### This is resampling
### But it increases the pitch of the audio while making it faster
### Decreasing sample rate will reduce the pitch
### This is called the Chipmunk Effect, which you would come across when you fast forward an audio tape.

```python
x, Fs = sf.read('D:\\v4\\sample.wav')

#print('The sample data is \n', x)
#print('\nThe sample rate is ', Fs )
#print('\nDuration of the audio is ', len(x)/Fs, ' seconds')
```

```python
sd.play(x, Fs)  # Normal
```

```python
sd.stop()
```

## <i> Deleting every n<sup>th</sup> sample data point</i>

```python
y = [x[i] for i in range(len(x)) if i%4 != 0]
sd.play(y, Fs)

#print('\nLength of y is ', len(y))
#print('\nDuration of this audio is ', len(y)/Fs, ' seconds')
```

### There are a lot of algorithms out there to perform audio scaling!
### The classic ones are the phase vocoders, SOLA, PSOLA etc...
### But what I have implemented is the Epoch Based Continuous Time and Pitch Scaling Algorithm!

## <u>About the Algorithm EBCTPS</u>

```python
import EBCTPS

epochs = EBCTPS.epoch(x, Fs)
time_scale = EBCTPS.ETS(x, Fs, 2, epochs)
pitch_scale = EBCTPS.EPS(x, Fs, 0.8, epochs)
sd.play(time_scale, Fs)
sd.wait()

#sd.play(pitch_scale, Fs)
#sd.wait()
```

```python
plt.plot(list(range(len(x))), x, 'r', list(range(len(time_scale))), time_scale, 'b')
#plt.show()
```

```python
plt.plot(list(range(len(x))), x, 'r', list(range(len(pitch_scale))), pitch_scale, 'b')
#plt.show()
```
